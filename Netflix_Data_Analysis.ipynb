## Exploratory Data Analysis (EDA)

### Q1 Does Netflix have more Movies or TV Shows?

```python
content_type = df['type'].value_counts()

content_type
```

```python
ax = content_type.plot(kind='bar', color=['red'])
ax.set_facecolor('black')
ax.figure.set_facecolor('black')

plt.title("Movies v/s TV Shows on Netflix", color='red')
plt.xlabel("Content type", color='white')
plt.ylabel("Number of Titles", color='white')
plt.xticks(rotation=0, color='white')
plt.yticks(color='white')

for index, value in enumerate(content_type):
    plt.text(index, value + 30, str(value), ha="center", fontsize=10, color='white')

plt.savefig("Images/Movies_vs_TVShows_chart.png", dpi=300, bbox_inches="tight", facecolor='black')

plt.show()
```

### Business Question 1:
1. Which type has more content?
Ans. Movies have more content in Netflix.
2. Approximately how many Movies are there?
Ans. Approximately there are 6131 Movies in Netflix
3. Approximately how many TV Shows are there?
Ans. Approximately there are 2676 TV Shows in Netflix.

### Q2 Which countries produce the most Netflix content?

```python
df['country'].head(10)
```

```python
country_count = df["country"].value_counts().head(10)

country_count
```

```python
ax = country_count.plot(kind='bar', figsize=(10,5), color='red')
ax.set_facecolor('black')
ax.figure.set_facecolor('black')

plt.title("Top 10 countries producing Netflix content", color='red')
plt.xlabel("Country", color='white')
plt.ylabel("Number of Titles", color='white')
plt.xticks(rotation=45, color='white')
plt.yticks(color='white')

for index, value in enumerate(country_count):
    plt.text(index, value + 20, str(value), ha="center", fontsize=9, color='white')

plt.savefig("Images/Top_10_countries_content.png", dpi=300, bbox_inches="tight", facecolor='black')

plt.show()
```

### Business Question 2:
1. Which country has the highest number of Netflix titles?
Ans. United States has highest number (2818) of Netflix titles.
2. Is India among the top producers?
Ans. Yes, India is in the top 10 list and it has secured 2nd Position next to the US.
3. Which five countries appear at the top?
Ans. US, India, UK, Japan, South Korea are the Top 5 countries
4. Why do you think these countries dominate Netflix's content library?
Ans. The United States, India, the United Kingdom, Japan, and South Korea have large entertainment industries, produce a high volume of movies and TV shows, and have strong partnerships with Netflix. These countries also have large audiences and export their content globally.

### Q3 Which content ratings are most common on Netflix?

```python
rating_count = df['rating'].value_counts()

rating_count
```

```python
ax = rating_count.plot(kind="bar", figsize=(10,5), color=[(1,0,0)])
ax.set_facecolor('black')
ax.figure.set_facecolor('black')

plt.title("Distrubtion of Netflix content Ratings", color='red')
plt.xlabel("Ratings", color='white')
plt.ylabel("Number of Titles", color='white')
plt.xticks(color='white')
plt.yticks(color='white')

for  index, value in enumerate(rating_count):
    plt.text(index, value + 20, str(value), ha="center", fontsize=8, color='white')

plt.savefig("Images/Netflix_content_Ratings.png", dpi=300, bbox_inches="tight", facecolor='black')

plt.show()
```

```python
rating_count.head(6).plot(kind="pie", autopct='%1.1f%%', figsize=(8,8), colors = ["#E50914", "#B20710", "#F40612", "#8B0000", "#FF6F61", "#FFC0CB"])

plt.xlabel("Top 6 Netflix content Rating")
plt.ylabel("") #Removes default y-axis label

plt.savefig("Images/Top_6_Netflix_content_Rating.png", dpi=300, bbox_inches="tight")

plt.show
```

### Business Question 3
1. Which rating is the most common on Netflix?
Ans. TV-MA is the most common rating with 3207 in Netflix.
2. What does that rating mean?
Ans. Rating is the one word description which indicates the content of the Movie or TV Show.
3. Which are the top 5 ratings?
Ans. TV-MA, TV-14, TV-PG, R, TV-13 are the top 5 Ratings in Netflix.
4. What does this tell us about Netflix's target audience?
Ans. With this analysis we conclude that TV-MA is the most common content rating, indicating that Netflix currently offers a large amount of mature-audience content. This suggests that adult viewers form a significant portion of Netflix's content strategy.

### Q4 What are the Top 10 Genres content on Netflix?

```python
genres_count = df["listed_in"].value_counts().head(10)

genres_count
```

```python
ax = genres_count.head(10).plot(kind="bar", figsize=(10,5), color='red')
ax.set_facecolor('black')
ax.figure.set_facecolor('black')

plt.title("Top 10 Genres", color="red")
plt.xlabel("Genres", color='white')
plt.ylabel("No. of Titles", color='white')
plt.xticks(rotation=15, color='white')
plt.yticks(color='white')

for index, value in enumerate(genres_count):
    plt.text(index, value + 5, str(value), ha='center', fontsize=10, color='white')

plt.savefig("Images/Top_10_Genres.png", dpi=300, bbox_inches="tight", facecolor='black')

plt.show()
```

### Q5 How has Netflix's content grown over the years?

```python
year_count = df["release_year"].value_counts().sort_index()

year_count
```

```python
ax = year_count.plot(figsize=(10,5), color='red')
ax.set_facecolor('black')
ax.figure.set_facecolor('black')

plt.title("Netflix content growth over the years", color='red')
plt.xlabel("Years", color='white')
plt.ylabel("No. of Releases", color='white')
plt.xticks(color='white')
plt.yticks(color='white')

plt.grid(True)

plt.savefig("Images/Netflix_content_over_years", dpi=300, bbox_inches='tight', facecolor='black')

plt.show()
```

### Business Question 5
1. Which year had the highest number of releases?
Ans. 2018 recorded the highest number of Netflix content releases, with 1,147 titles.
2. Did Netflix's content generally increase over time?
Ans. Yes. Netflix's content releases generally increased over time. The growth accelerated after 2000 and became particularly strong after 2015.
3. What trend do you observe?
Ans. The number of titles released increased gradually until the early 2000s and grew rapidly after 2015, reaching a peak in 2018. The apparent decline after 2020 is likely due to the dataset covering only part of 2021, rather than indicating a true decrease in Netflix's content production.
